Дано файл $f$, який містить відомості про іграшки: вказується назва
іграшки (наприклад, лялька, кубики, м$'$яч, конструктор і т. д.), її
вартість в копійках і можливі вікові межі дітей, для яких іграшка
призначена (наприклад, для дітей від двох до п$`$яти років). Отримати
ціни всіх кубиків оформленні по зразку ...грн ...коп.

```python
with open('20_f.txt', encoding = 'utf-8') as f:
    data_f = f.read(); f.close()
list_f = data_f.split('\n')
list_f = [list_f[i].split(', ') for i in range(len(list_f))]
print(data_f, "\n\n", [f"{s[1][:-2]} грн {s[1][-2:]} коп"
                       for s in list_f if ("кубики" in s[0])])
```
