### Завдання № 368
Дано дійсні числа $a_1,...,a_{10}$, $b_1,...,b_{20}$. Отримати
дійсну матрицю $[c_{ij}]_{i=1,...,20; j=1,...,10}$, для якої
$c_{ij}=a_j/(1+|b_i|)$.

```python
from random import randint

a = [randint(-9, 9) for i in range(10)]
b = [randint(-9, 9) for j in range(20)]
rez = [[round(j / (1 + abs(i)), 1) for j in a] for i in b]
print(f'a = {a}\nb = {b}\n\nMatrix')
for i in rez: print(i)
```
