### Завдання № 700(б)
Дана матриця $A$ порядку $n$. Отримати матрицю $AB$; елементи
матриці $B$ визначаються по формулі:
$b_{ij}=\begin{cases}
   \frac{1}{i+j-1}, &\text{якщо } i \leq j \\
   \frac{1}{i+j+1} &\text{в протилежному випадку }
\end{cases}$

```python
from random import randint
from numpy import matrix

try:
    n = int(input("Enter the order of the matrix:"))
    if n > 0:
        a = matrix([[randint(-9, 9) for _ in range(n)] for _ in range(n)])
        b = matrix([[round(1 / (i + j - 1), 1) if i <= j else round(1 / i + j + 1, 1)
                     for j in range(1, n + 1)] for i in range(1, n + 1)])
        print(f"Matrix A:\n{a}\nMatrix B:\n{b}\nMatrix AB:\n{a * b}")
    else:
        print("Input error!")
except ValueError:
    print("Input error!")
```
