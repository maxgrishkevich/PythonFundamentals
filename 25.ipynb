Дано дійсні числа $a_1, ..., a_n$, дійсна квадратна матриця порядку
$n (n \geq 6)$. Отримати дійсну матрицю розміру $n \times (n+1)$,
вставивши у початкову матрицю між п$`$ятим та шостим стовбцем новий
стовбець з елементами $a_1, ..., a_n$.

```python
from random import randint
from numpy import matrix

try:
    n = int(input("Enter the order of the matrix:"))
    if n >= 6:
        a = [randint(-9, 9) for _ in range(n)]
        m = [[randint(-9, 9) for _ in range(n)] for _ in range(n)]
        print(f"Row a:\n{a}\nMatrix:\n{matrix(m)}")
        for i in range(n): m[i].insert(5, a[i])
        print(f"New matrix:\n{matrix(m)}")
    else:
        print("Input error!")
except ValueError:
    print("Input error!")
```
